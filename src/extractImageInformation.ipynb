Display images

```python
%matplotlib inline

import tifffile as tiff
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

image_paths = [
    "/../scratch/confocal_cropped.tif",
    "/../scratch/confocal.tif",
    "/../scratch/cortical.tif",
    "/../scratch/cortical_cropped.tif"
]

# Initialize a figure to plot multiple images
plt.figure(figsize=(12, 10))

# Loop through each image path
for i, image_path in enumerate(image_paths):
    # Load the TIFF image
    image = tiff.imread(image_path)
    
    # Select the middle slice along the first axis
    slice_index = image.shape[0] // 2
    selected_slice = image[slice_index, :, :]
    
    # Plot the selected slice
    plt.subplot(2, 2, i + 1)
    plt.imshow(selected_slice, cmap='gray')
    plt.title(f'{image_path} - Slice {slice_index}')
    plt.colorbar()

# Adjust layout and display the figure
plt.tight_layout()
plt.show()

```
